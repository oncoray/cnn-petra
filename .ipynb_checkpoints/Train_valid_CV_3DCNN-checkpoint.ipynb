```python
# create a data frame that contains endpoint information and patient id. We need patient id for:
# 1- startification, 2- To keep track of included patients in train and valid folds
data_org= pd.read_csv(r'D:\Transcend_(E)\_IramS\PETra_ML_DL_Analysis\Processed_Features\PET_Features_Training\PET_PETra_Clinical_Imaging_Features_Robust.csv')

# Provide the path to folder with 3D patches stored in nifti format
data_org['Path']= 'D:\Transcend_(E)\_IramS\PETra_ML_DL_Analysis\Deep_learning\Data\PETra_Cox_model_data'+'\\'+ data_org.subject_id+'_0_base_img.nii.gz'
data_org = data_org[['subject_id','LC','LCtime','Path']] # LC = event status, LCtime = time to event 
data= data_org.rename(columns={'LC': 'status', 'LCtime': 'time'}) # rename the columns time and status column for easy interprestation in remaining script 
X=data[['subject_id']] # get pateint ids for startification 
y=data[['status']] # get labels for startification
```

```python
'''
Image pre-processing 
'''

def read_nifti_file(filepath):
    """Read and load volume"""
    # Read file
    scan = nib.load(filepath)
    # Get raw data
    scan = scan.get_fdata()
    return scan

def normalize(volume):
    """Normalize the volume for PET data"""
    volume[volume > 10] = 10
    volume = volume/10
    volume = volume.astype("float32")
    return volume

def process_scan(path):
    """Read and resize volume"""
    # Read scan
    volume = read_nifti_file(path)
    # Normalize
    volume = normalize(volume)    
    return volume
```

```python
'''
pre-processing the batch data on fly using data augmentation and adding channel
data augmentation using batchgenerator library is imported from Aug3D.py
'''

def train_preprocessing(volume, label):
    """Process training data by augmenting and adding a channel."""
    volume = Aug3D.augment_data(volume)
    volume = tf.squeeze(volume)
    volume = tf.expand_dims(volume, axis=3)
    volume = tf.reshape(volume, shape=(60, 60,44,1))
    return volume, label


def validation_preprocessing(volume, label):
    """Process validation data by only adding a channel."""
    volume = tf.expand_dims(volume, axis=3)
    return volume, label
```

```python
'''
tf data_loader function peform train and validation augmentation on fly
'''
def data_loaders(x_train,y_train,x_val,y_val):
    train_loader = tf.data.Dataset.from_tensor_slices((x_train, y_train))
    validation_loader = tf.data.Dataset.from_tensor_slices((x_val, y_val))

    batch_size = 16
    # Augment the on the fly during training.
    train_dataset = (
        train_loader.shuffle(len(x_train))
        .map(train_preprocessing)
        .batch(batch_size)
        .prefetch(2)
    )

    # Only rescale.
    validation_dataset = (
        validation_loader.shuffle(len(x_val))
        .map(validation_preprocessing)
        .batch(batch_size)
        .prefetch(2)
    )
    
    return train_dataset,validation_dataset
```

```python
def cox_log_likelihood(labels, risk):
    """
    The negative of the cox partial log-likelihood (since in DL one
    typically minimizes a loss instead of maximizing likelihood).
    Parameters
    ----------
    labels: 2d numpy array (n_samples x 2)
        first column has events status, second column time
    This does not assume that data is ordered by decreasing
    survival time but does it beforehand.
    
    Method taken from: https://github.com/oncoray/cnn-hnscc
    """
  
    e = labels[:, 0] # first column in my data is event status
    t = labels[:, 1]

    # sort data
    t_ordered, idx = tf.nn.top_k(t, k=tf.size(t))
    e_ordered = tf.gather(e, indices=idx, axis=0)
    risk_ordered = tf.gather(risk, indices=idx, axis=0)

    # compute likelihood
    sum_risk = tf.reduce_sum(e_ordered * risk_ordered)
    log_sums = tf.math.log(tf.cumsum(tf.exp(risk_ordered)))
    log_sums = tf.reduce_sum(e_ordered * log_sums)

    lcpl = sum_risk - log_sums
    return lcpl


def neg_cox_log_likelihood(labels, risk):
    return -1. * cox_log_likelihood(labels, risk)

```

```python
'''
function to fit model with given hyperparameters
'''
def model_fitting(model, initial_learning_rate,decay_steps, decay_rate, epochs,train_dataset,validation_dataset,fold,weights):# Compile model.
# if cudnn error occurs in model compile, detach and attach environment again 
    initial_learning_rate = initial_learning_rate
    lr_schedule = keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate, decay_steps=decay_steps, decay_rate=decay_rate, staircase=True
    )
    model.compile(
        loss=neg_cox_log_likelihood,
        optimizer=keras.optimizers.Adam(learning_rate=lr_schedule,amsgrad=False),
    )

    # Define callbacks.
    #https://www.pyimagesearch.com/2021/06/30/how-to-use-the-modelcheckpoint-callback-with-keras-and-tensorflow/
    # save the trained model at the end of each fold
    fname = "3d_image_cox_reg_"+ str(fold) + ".hdf5"
    fname=os.path.join(weights,fname)
    checkpoint_cb = keras.callbacks.ModelCheckpoint(fname, save_best_only=True, mode="min",verbose=1)
    #checkpoint_cb = keras.callbacks.ModelCheckpoint(fname, mode="min",verbose=1, period=epochs) # if want to save model at the end of all epochs
    early_stopping_cb = keras.callbacks.EarlyStopping(monitor="val_loss", patience=2)
    
    # Train the model, doing validation at the end of each epoch
    model.fit(
        train_dataset,
        validation_data=validation_dataset,
        epochs=epochs,
        shuffle=True,
        verbose=2,
        #callbacks=[checkpoint_cb, early_stopping_cb, tensorboard_callback], #use tensorboard when performance visualization is required 
        callbacks=[checkpoint_cb, early_stopping_cb], 
    )
    return model
```

```python
'''
Once the model is trained it is applied to train and vlaidation data to collect model predictions using evaluate function 
'''
def evaluate(model, x_train, x_valid, fold, weights):
    fname = "3d_image_cox_reg_"+ str(fold) + ".hdf5"
    fname=os.path.join(weights,fname)
    model.load_weights(fname)

    prediction_train = []
    prediction_valid = []
    for i in range(0,len(x_train)):
        prediction = model.predict(np.expand_dims(np.expand_dims(x_train[i], axis=3), axis=0))[0]
        prediction_train.append(prediction[0])


    for i in range(0,len(x_valid)):
        prediction = model.predict(np.expand_dims(np.expand_dims(x_valid[i], axis=3), axis=0))[0]
        prediction_valid.append(prediction[0])
    return prediction_train, prediction_valid
```

```python
'''Compute concordance index'''
def compute_ci(y_train, yhat_train, y_valid, yhat_valid):
    CI_train  = concordance_index(y_train[:,1], yhat_train,y_train[:,0])
    CI_valid  = concordance_index(y_valid[:,1], yhat_valid,y_valid[:,0])
    
    return CI_train, CI_valid
```

```python
i=1
splits=2 # number of splits for Cross-validation
repeats=2 # number of repeats for Cross-validation
kfold  = RepeatedStratifiedKFold(n_splits=splits,n_repeats=repeats,random_state=0) # create k-folds
CI_train_all=[] # create empty list to store all concordance index train
CI_valid_all=[] # create empty list to store all concordance index valid

# here we will store all 
risk_ensemble_train =pd.DataFrame()
risk_ensemble_train['subject_id'] = X.subject_id # store all subject ids, 
risk_ensemble_valid =pd.DataFrame()
risk_ensemble_valid['subject_id'] = X.subject_id

for train, valid in kfold.split(X, y): #train and valid contain index position of splits created 
    X_train, X_valid = X.iloc[train.tolist()], X.iloc[valid.tolist()] #get train and valid IDs  
    X_train_subvolumes = data.loc[data.subject_id.isin(np.array(X_train.subject_id))] 
    X_valid_subvolumes =  data.loc[data.subject_id.isin(np.array(X_valid.subject_id))]
    x_train = np.array([process_scan(path) for path in X_train_subvolumes.Path])
    y_train = np.array(X_train_subvolumes[['status','time']])
    x_val = np.array([process_scan(path) for path in X_valid_subvolumes.Path])
    y_val = np.array(X_valid_subvolumes[['status','time']])
    
    train_dataset,validation_dataset = data_loaders(x_train,y_train,x_val,y_val)
    
    ''' Three models can be used here
    1- DenseNet3D imported from DenseNet3D.py
    2- ResNet3D imported from ResNet3D.py
    3- Vgg3D imported from Vgg3D.py file
    comment models that are not required
    '''
    
    #model =Vgg3D.Vgg3D_3blocks(input_shape=(60,60,44,1))
    #model =ResNet3D.Resnet3DBuilder.resnet_v3((60,60,44,1),1,reg_factor=1e-4)
    model = DenseNet3D.DenseNet3DImageNet121(input_shape=(60,60,44,1), include_top=False)
    
    model = Model(inputs=model.input, outputs=model.output)
    print("########### Model fitting for fold %d of %d ###########"% (i,(splits*repeats)))
    
    #Model hyperparameters
    trained_model= model_fitting(model,initial_learning_rate= 0.0001,decay_steps=1000,decay_rate=0.96,epochs=2,
                                 train_dataset=train_dataset,validation_dataset=validation_dataset,fold=i,weights=weights)
    
    yhat_train, yhat_valid = evaluate(trained_model, x_train, x_val,i,weights)
    
    
    
    ##################### collect Subject_risk in current fold for computing Ensemble risk ##############################
    
    '''Train'''
    for s_id_train, yhat_t in zip(np.array(X_train_subvolumes.subject_id), np.array(yhat_train)):
        # X_train subvolumes have subject ids selected in current fold, These id are matched with risk_ensemble_train dataframe 
        # where id is matched, the prediction is placed in front of it for each cv fold 
            index = risk_ensemble_train.index[risk_ensemble_train['subject_id'] == s_id_train].tolist() 
            risk_ensemble_train.loc[index[0], 'risk_fold_'+str(i)] = yhat_t
    '''Valid'''
    for s_id_valid, yhat_v in zip(np.array(X_valid_subvolumes.subject_id), np.array(yhat_valid)):
            index = risk_ensemble_valid.index[risk_ensemble_valid['subject_id'] == s_id_valid].tolist()
            risk_ensemble_valid.loc[index[0], 'risk_fold_'+str(i)] = yhat_v
            
    ci_train_all, ci_valid_all = compute_ci(y_train,yhat_train,y_val,yhat_valid)
    CI_train_all.append(ci_train_all)
    CI_valid_all.append(ci_valid_all)
    print('########## CI Train at fold %d is %.2f ############' % (i,(ci_train_all)))
    print('########## CI Valid at fold %d is %.2f ############' % (i,(ci_valid_all)))
    i=i+1
```

```python
#Calculate the ensemble risk for train and validation 
risk_ensemble_train.fillna(0)
risk_ensemble_train['Ensemble'] = risk_ensemble_train.loc[:,'risk_fold_1': 'risk_fold_'+str(splits*repeats)].mean(axis=1)
risk_ensemble_train.fillna(0)
risk_ensemble_valid['Ensemble'] = risk_ensemble_valid.loc[:,'risk_fold_1': 'risk_fold_'+str(splits*repeats)].mean(axis=1)
#Calculate CI from ensemble risk 
risk_ensemble_train['status'],risk_ensemble_train['time'] = data['status'],data['time']
risk_ensemble_valid['status'],risk_ensemble_valid['time'] = data['status'],data['time']
CI_ensemble_train  = concordance_index(risk_ensemble_train.time,risk_ensemble_train.Ensemble,risk_ensemble_train.status)
CI_ensemble_valid  = concordance_index(risk_ensemble_valid.time,risk_ensemble_valid.Ensemble,risk_ensemble_valid.status)
```

```python
def LogRank_test(risk_ensemble):
    pred= logrank_test(np.array(risk_ensemble.loc[risk_ensemble.risk_group==0].time),
                       np.array(risk_ensemble.loc[risk_ensemble.risk_group==1].time),
                       np.array(risk_ensemble.loc[risk_ensemble.risk_group==0].status),
                       np.array(risk_ensemble.loc[risk_ensemble.risk_group==1].status),
                       alpha=0.95)
    p_value = pred.p_value
    return p_value
```

```python
# get optimal cutoff and p-value for train data using log-rank test, optimal cutoff is one where p-value is minimum  
p_value = list()
cut_off = list()

for i in range(0,risk_ensemble_train.shape[0]):
    cut = risk_ensemble_train.Ensemble[i]
    risk_ensemble_train['risk_group'] = np.where(risk_ensemble_train['Ensemble'] > cut, 1,0)
    p_val= LogRank_test(risk_ensemble_train)
    p_value.append(p_val)
    cut_off.append(cut)
    
optimal_cutoff= cut_off[p_value.index(min(p_value))]
optimal_p_value = min(p_value)
```

```python
#Valid p-values at optimal threshold using log-rank test 
risk_ensemble_train['risk_group'] = np.where(risk_ensemble_train['Ensemble'] > optimal_cutoff, 1,0)
risk_ensemble_valid['risk_group'] = np.where(risk_ensemble_valid['Ensemble'] > optimal_cutoff, 1,0)
p_val_valid= LogRank_test(risk_ensemble_valid)

# Final sheet with train and validation predictions for 25 CV folds
risk_ensemble_train.to_excel('Ensemble_train_risk_at_optimal_cutoff.xlsx', index=None, header=True)
risk_ensemble_valid.to_excel('Ensemble_valid_risk_at_optimal_cutoff.xlsx', index=None, header=True)
```

```python
#Display final KM plots based on train and validation ensemble predictions 
#plot KM
size=16
data_train= risk_ensemble_train
data_val= risk_ensemble_valid
ix= data_train["risk_group"] ==1
ix2= data_val["risk_group"] ==1
#plotting images#
plt.rc('xtick', labelsize=size)    # fontsize of the tick labels
plt.rc('ytick', labelsize=size)
fig,ax = plt.subplots(nrows=1, ncols=2,figsize= (14,7))

#training kaplan meier
kmf_train_high = KaplanMeierFitter()
ax[0] = kmf_train_high.fit(data_train.loc[ix]["time"], data_train.loc[ix]['status'], label='High risk').plot(ax=ax[0],
                            show_censors=True,ci_show=False, xlim=(0,60), color= "blue",yticks=(0.0,0.2,0.4,0.6,0.8,1),
                            ylim=(0.0,1),xticks=(0,12,24,36,48,60))

kmf_train_low = KaplanMeierFitter()
ax[0] = kmf_train_low.fit(data_train.loc[~ix]["time"], data_train.loc[~ix]['status'], label='Low risk').plot(ax=ax[0],
                             show_censors=True,ci_show=False, xlim=(0,60),color="orange",yticks=(0.0,0.2,0.4,0.6,0.8,1),
                             ylim=(0.0,1),xticks=(0,12,24,36,48,60))


#validation kaplan-meier
kmf_val_high = KaplanMeierFitter()
ax[1]= kmf_val_high.fit(data_val.loc[ix2]["time"], data_val.loc[ix2]['status'], label='High risk').plot(ax=ax[1], show_censors=True,ci_show=False, xlim=(0,60), color= "blue",yticks=(0.0,0.2,0.4,0.6,0.8,1), ylim=(0.0,1),xticks=(0,12,24,36,48,60))

kmf_val_low = KaplanMeierFitter()
ax[1] = kmf_val_low.fit(data_val.loc[~ix2]["time"], data_val.loc[~ix2]['status'], label='Low risk').plot(ax=ax[1], show_censors=True,ci_show=False, xlim=(0,60),color="orange",yticks=(0.0,0.2,0.4,0.6,0.8,1), ylim=(0.0,1),xticks=(0,12,24,36,48,60))
#settings for training km
ax[0].set_xlabel("Time [months]",size=size)
ax[0].set_ylabel("Local Recurrence",size=size)
ax[0].text(45, 0.25, "p="+str(round(optimal_p_value,3)), size=size)
#settings for validation km
ax[1].set_xlabel("Time [months]",size=size)
ax[1].text(45, 0.25, "p="+str(round(p_val_valid,3)), size=size)



ax[0].tick_params(axis='both', which='major', labelsize=size)
ax[1].tick_params(axis='both', which='major', labelsize=size)
ax[0].legend(fontsize=size, loc=1)
ax[1].legend(fontsize=size, loc=1)
ax[0].set_title("Training", size=size)
ax[1].set_title("Internal Validation",size=size)

from lifelines.plotting import add_at_risk_counts
util.add_at_risk(kmf_train_high, kmf_train_low, ax=ax[0],rows_to_show=['At risk'])
util.add_at_risk(kmf_val_high, kmf_val_low, ax=ax[1],labels=["", ""],rows_to_show=['At risk'])
plt.tight_layout()
plt.savefig("Train_Valid_starta.png", dpi=300)
#dpi stands for dots per pixel per inch, good quality is above 300
```
